```python
import numpy as np
import pandas as pd

DF = pd.read_csv("./chicago-taxi-rides.csv")
DF = DF.dropna(how='any')
DF.iloc[:, 0] = DF.iloc[:, 0].astype(np.int)

G = np.zeros((77, 77))
for i in range(len(DF)):
    G[DF.iloc[i, 0] - 1][DF.iloc[i, 1] - 1] += DF.iloc[i, 2]
```

```python
N = 77
beta = 0.85
iterations = 100

rank_old = np.ones((1, 77))
rank_new = np.zeros((1, 77))

for i in range(N):
    rank_old[0][i] = rank_old[0][i] / sum(G[i])

iter_rank = dict()

parent = dict()
for i in range(N):
    parent[i] = []

for i in range(N):
    for j in range(N):
        if G[i][j] != 0:
            parent[j].append(i)


for i in range(2, iterations+1):
    if (i % 2 == 0):
        iter_rank[i] = None

for _ in range(1, iterations):
    print("[INFO]. Iterations : ", _)
    print("[INFO]. Old Rank   : ", rank_old[0][:20])
    rank_dash_new = np.zeros((1, 77))
    rank_new = np.zeros((1, 77))
    for j in range(N):
        if len(parent[j]) != 0:
            rank_dash_new[0][j] = sum([beta * (rank_old[0][i] / sum(G[i])) for i in range(N)])
        
        rank_new[0][j] = rank_dash_new[0][j] + ((1 - sum(rank_dash_new[0])) / N)
    
    print("[INFO]. New Rank   : ", rank_new[0][:20])
    rank_old = rank_new

    if _ % 2 == 0:
        iter_rank[_] = rank_old

    al = input()
```
